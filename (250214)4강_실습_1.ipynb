# **[텐서플로에서 실행하는 선형회귀]**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 텐서플로의 케라스 API에서 필요한 함수들 불러오기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

model = Sequential()
```

가설 함수는 H(x) = wx + b 이다.


1. 이때 출력되는 값(=성적)이 하나씩 이므로 Dense함수의 첫 번째 인자에 1이라고 설정한다.
2. 입력될 변수(=학습 시간)도 하나뿐이므로 input_dim역시 1이라고 설정한다.
3. 입력된 값을 다음층으로 넘길 때 각 값을 활성화 함수에 따라 처리하는데, activation은 활성화 함수를 정하는 옵션이다.
4. 여기에서는 선형회귀를 다루고 있으므로 'linear'라고 적어주면 된다.







```python
# 출력 값, 입력 변수, 분석 방법에 맞게끔 모델 설정
model.add(Dense(1,input_dim=1, activation='linear'))
```

1. 앞서 배운 경사하강법을 실행하려면 옵티마이저에 sgd라고 설정한다.
2. 손실함수는 평균 제곱 오차를 사용할 것이므로 mse라고 설정한다.

```python
# 오차 수정을 위해 경사 하강법(sgd)을 , 오차의 정도를 판단하기 위해 평균 제곱 오차(mse)사용
model.compile(optimizer='sgd', loss='mse')
```

```python
# 오차를 최소화하는 과정을 2000번 반복한다
model.fit(x,y,epochs=2000)

plt.scatter(x,y)
plt.plot(x,model.predict(x),'r')
plt.show()
```

```python
# 임의의 시간을 집어넣어 점수를 예측하는 모델을 테스트 해보기
hour = 7
prediction = model.predict([[hour]])
print("%.f시간을 공부할 경우의 예상점수는 %.02f점입니다."%(hour, prediction))
```
